# Querys ejercicio # 2

1. Nombres y créditos de clientes de los compradores de CEMACO ordenado por crédito descendente y nombre ascendente.

```python
%sql SELECT cliente.nombre, cliente.credito FROM cliente INNER JOIN compra ON cliente.idCliente = compra.idCliente WHERE compra.idTienda = "375836"
```

2. Agregar en el resultado anterior el crédito en dólares a un tipo de cambio de 7.8.

```python
%sql SELECT cliente.nombre, cliente.credito*7.8 FROM cliente INNER JOIN compra ON cliente.idCliente = compra.idCliente WHERE compra.idTienda = "375836"
```

3. Parejas de clientes con el mismo crédito que no sean ellos mismos y que no haya parejas duplicadas

```python
%sql SELECT DISTINCT cliente.nombre AS cliente1, c.nombre AS cliente2, cliente.credito FROM cliente INNER JOIN (SELECT * FROM cliente) AS c ON cliente.credito = c.credito WHERE cliente.nombre != c.nombre;
```

 4. Clientes que han comprado en CEMACO y no en PAIZ

```python
%%sql SELECT cliente.idCliente, cliente.nombre FROM cliente INNER JOIN compra ON cliente.idCliente = compra.idCliente WHERE compra.idTienda = "375836" AND NOT EXISTS
(SELECT cliente.nombre FROM cliente  INNER JOIN compra ON cliente.idCliente = compra.idCliente WHERE compra.idTienda = "209755");
```

5. Todas las tiendas que tengan otras tiendas en la misma dirección

6. Clientes con los créditos más altos (sin utilizar MAX())
